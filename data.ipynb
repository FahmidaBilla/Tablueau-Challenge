```python
import pandas as pd
import numpy as np
import glob
```

```python
# list all csv files only
csv_files = glob.glob('Resources/JC-*.csv')
csv_files
```

```python
combined_citibike_csv = pd.concat([pd.read_csv(file) for file in csv_files ], ignore_index=True)
combined_citibike_csv
```

```python
combined_citibike_csv.dtypes
```

```python
combined_citibike_csv["started_at"] = pd.to_datetime(combined_citibike_csv["started_at"])
combined_citibike_csv["ended_at"] = pd.to_datetime(combined_citibike_csv["ended_at"])

combined_citibike_csv.dtypes
```

```python
combined_citibike_csv = combined_citibike_csv.sort_values("started_at")

combined_citibike_csv.head()
```

```python
#break out month
combined_citibike_csv['Trip Month'] = combined_citibike_csv["started_at"].dt.month_name()
combined_citibike_csv.head()
```

```python
combined_citibike_csv.value_counts("member_casual")
```

```python
combined_citibike_csv.value_counts("Trip Month")
```

```python
# Export the DataFrame as a CSV file. 
combined_citibike_csv.to_csv("Resources/combined_citibike.csv", encoding='utf8', index=False)
```
